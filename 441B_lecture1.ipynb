```
x_axis = df1['FY']
y_axis = df1['Total_Revenue']


plt.figure(figsize=(10,6))
plt.bar(x_axis,y_axis,color="red",width=0.4)
plt.xlabel('Fiscal Year')
plt.ylabel('Total Revenue in Rupee Crores')
plt.title("Total Revenue of Delhi NCT")

plt.show()
```
